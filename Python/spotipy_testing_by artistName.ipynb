```python
# Search for artists whose names start with "A"
artist_name_list =[]
artist_id_list = []
artist_genres = []
artist_popularity =[]
artist_followers_count = []
limit = 50 
offset =0
while True:
    results = sp.search(q='artist:A*', type='artist', limit=50, offset=offset)
    artists = results['artists']['items']
    # Extract artist IDs and make API calls for each artist
    for artist in artists:
        artist_id = artist['id']
        artist_info = sp.artist(artist_id)
        # Process artist information as needed
        print(artist['name'], artist['popularity'], artist['genres'])
        # add to the artist_name_list 
        artist_name_list.append(artist['name'])
        artist_followers_count.append(artist['followers']['total'])
        artist_popularity.append(artist['popularity'])
        artist_id_list.append(artist['id'])

    # Check if there are more tracks to retrieve
    if len(artists) < limit:
        break
    
    offset += limit


  
   
```
