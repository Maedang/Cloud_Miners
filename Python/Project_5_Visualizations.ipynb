```python
query_2 = pd.read_csv("Resources/csv/query_2.csv")
query_2
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Scatter plot
plt.scatter(query_2['release_year'], query_2['avg_duration'], color='black', label='Data Points')

# Fitted curve
degree = 2  # Degree of the polynomial curve
coeffs = np.polyfit(query_2['release_year'], query_2['avg_duration'], degree)
curve = np.poly1d(coeffs)
x_range = np.linspace(query_2['release_year'].min(), query_2['release_year'].max(), 100)  # Generate a range of x-values for the curve
plt.plot(x_range, curve(x_range), color='blue', label='Fitted Curve')

# Plot settings
plt.xlabel('Release Year')
plt.ylabel('Average Duration')
plt.title('Change of Average Duration over the Years')

# Display the plot
plt.legend()
plt.show()

plt.savefig('Duration Over the Year.jpg', format='jpeg')
```

```python
track_df = pd.read_csv("Resources/csv/tracks.csv")
track_df
```

```python
chart_df = pd.read_csv("Resources/csv/charts.csv")
chart_df
```

```python
release_df = pd.read_csv("Resources/csv/release.csv")
release_df
```

```python
merge_df1 = pd.merge(chart_df, release_df, on = "Track_Id", how= "outer")
merge_df1
```

```python
artist_df= pd.read_csv("Resources/csv/artists.csv")
artist_df
```

```python
merge_artist = pd.merge(merge_df1, artist_df, on = "Artist_Id", how = "inner")
merge_artist
```

```python
merge_track = pd.merge(merge_df1, track_df, on= "Track_Id", how = "inner")
merge_track
```

```python
#What is the popular artist name 
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
artist_name = " ".join(review for review in merge_artist.Artist_Name.astype(str))
print(len(artist_name))
print ("There are {} words in the combination of all cells in column artist name.".format(len(artist_name)))
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white", width=1000, height=600,collocations=False).generate(artist_name)
plt.axis("off")
#plt.figure(figsize=(20,10))
plt.tight_layout(pad=0)
plt.imshow(wordcloud, interpolation='bilinear')
plt.show()
```

```python
#What is the most popular song name 
track_name = " ".join(review for review in merge_track.Track_Name.astype(str))
print(len(track_name))
print ("There are {} words in the combination of all cells in column track name.".format(len(track_name)))
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="white", width=1000, height=600,collocations=False).generate(track_name)
plt.axis("off")
#plt.figure(figsize=(20,10))
plt.tight_layout(pad=0)
plt.imshow(wordcloud, interpolation='bilinear')
plt.show()
```
